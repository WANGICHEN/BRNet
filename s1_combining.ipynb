```python
ff_pair = pd.read_csv('file_pairs.csv').values.tolist()
print(len(ff_pair))

# 合併並壓縮保存為MAT檔案
for f in ff_pair:
    print(f[0])
    pid = basename(dirname(f[0]))
    mat_open = mat73.loadmat(f[0])
    mat_close = mat73.loadmat(f[1])
    
    combined_data = {}
    
    for key in mat_open.keys():
        try:
            combined_data[key + '_open'] = mat_open[key]
        except TypeError:
            print(f"Could not convert {key} in {f[0]} to array")
    
    for key in mat_close.keys():
        try:
            combined_data[key + '_close'] = mat_close[key]
        except TypeError:
            print(f"Could not convert {key} in {f[1]} to array")
    
    output_filename = pid + '_combined.mat'
    savemat(output_filename, combined_data, do_compression=True)
    print(f'Saved combined data to {output_filename}')
    
```
